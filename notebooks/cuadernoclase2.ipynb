# Veamos el ejercicio de los polinomios en $\mathbb{R}^3$

La clase de abajo tiene dos métodos de evalauar, la primera dada por Ángel y la otra una propuesta adicional.
Además verifica elementos de longitud 4

- Finalmente, crear un procedimiento para asegurar que la lista ingresada tenga longitud 4

```python
class P3:
    def __init__(self, L):
        self.L = self.verificar_longitud(L)

    def verificar_longitud(self, L1):
        if len(L1) != 4:
            raise ValueError('La lista debe tener 4 elementos')
        return L1
    
    def evaluar(self, x):
        exp = 0
        result = 0
        for i in self.L:
            result += i*(x**exp)
            exp += 1
        return result

    def evaluar1(self, x):
        return sum([self.L[i]*(x**i) for i in range(len(self.L))])
```

```python
# Invocar el método super()
class Cuenta1(Cuenta):
    def __init__(self, nombre, saldo):
        super().__init__(nombre, saldo)
        self.numero_cuenta = self.numero_cuenta()

    def numero_cuenta(self):
        import random
        Letras = 'ABCDEFGHIJKLMOPQRSTUVWXYZ'
        Digitos = [str(i) for i in range(10)]
        L = random.sample(Letras, k = 4)
        D = random.choices(Digitos, k = 4)
        return ''.join(L + D)
 

    def depositar(self, cantidad):
        self.saldo += cantidad
        return f'Su nuevo saldo es {self.saldo}'
    
```

```python
C3.nombre, C3.saldo, C3.numero_cuenta
```

```python
class Cuenta2(Cuenta1):
    def numero_cuenta(self):
        import random
        letras = 'ABCDEFGHIJKLMONPQRSTUVWXYZ'
        return super().numero_cuenta() + ''.join(random.choices(letras, k= 2))
```

## Un diccionario básico

```python
class DiccionarioSecuencial:
    def __init__(self):
        # Lista de tuplas (clave, valor)
        self._data = []
    
    def __getitem__(self, key):
        """
        Retorna el valor asociado a la clave.
        Debe buscar de forma secuencial.
        Permite la sintaxis D[key]
        """
        pass
    
    def __setitem__(self, key, value):
        """
        Inserta o actualiza un par (clave, valor).
        Si la clave existe, actualiza su valor.
        Si no, la agrega al final.
        Permite la sintaxis D[key] = value
        """
        for i in range(len(self._data)):
            if self._data[i][0] == key:
                self._data.pop(i)
                break
        self._data.append((key, value))
    
    def __delitem__(self, key):
        """
        Elimina el elemento asociado a la clave.
        Si no existe, lanzar KeyError.
        del D[key]
        """
        pass
    
    def __iter__(self):
        """
        Retorna un iterador sobre las claves.
        """
        pass
    
    def __len__(self):
        """
        Retorna la cantidad de elementos.
        """
        pass
```
