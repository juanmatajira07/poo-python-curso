## Algo de fraccionarios

```python
from math import gcd
class Fraccion:
    
    def __init__(self, numerador, denominador):
        """
        Inicializa una fracciÃ³n.
        """
        self.numerador = numerador
        self.denominador = denominador

    def _simplificar(self):
        h = gcd(self.numerador, self.denominador)
        return Fraccion(self.numerador // h, self.denominador // h)

    
    def __str__(self):
        """
        Retorna una representación legible de la fracción.
        """
        return self.__repr__()
    
    def __repr__(self):
        """
        Retorna una representación técnica de la fracción.
        """
        return f'{self.numerador}/{self.denominador}'
    
    def __add__(self, other):
        """
        Retorna la suma de dos fracciones.
        """
        numerador = self.numerador*other.denominador + other.numerador*self.denominador
        denominador = self.denominador*other.denominador
        return Fraccion(numerador, denominador)._simplificar()
    
    def __sub__(self, other):
        """
        Retorna la resta de dos fracciones.
        """
        '''numerador = self.numerador*other.denominador - other.numerador*self.denominador
        denominador = self.denominador*other.denominador
        return Fraccion(numerador, denominador)'''
        return self + Fraccion(-other.numerador, other.denominador)        
    
    def __mul__(self, other):
        """
        Retorna la multiplicaciÃ³n de dos fracciones.
        """
        return Fraccion(self.numerador*other.numerador, self.denominador*other.denominador)._simplificar()
    
    def __truediv__(self, other):
        """
        Retorna la divisiÃ³n de dos fracciones.
        """
        return self*Fraccion(other.denominador, other.numerador)
    
```

```python
F1, F2 = Fraccion(120,85), Fraccion(2,2)
F1 + F2, F1 - F2, F1*F2, F1/F2
```

```python
class EstudianteV3:
    def __init__(self, nombre, promedio):
        self.nombre   = nombre
        self.__promedio = promedio

    def __repr__(self):
        return f'EstudianteV3 | nombre: {self.nombre} | promedio: {self.__promedio}'
    
    def get_promedio(self):
        return self.__promedio

    def set_promedio(self, nuevo_promedio):
        self.__promedio = nuevo_promedio
```

```python
class EstudianteV4:
    def __init__(self, nombre, promedio):
        self.__nombre   = nombre
        self.__promedio = promedio

    @property
    def nombre(self):
        return self.__nombre

    @nombre.setter
    def nombre(self, valor):
        if not isinstance(valor, str) or not valor.strip():
            raise ValueError("El nombre debe ser una cadena no vacía.")
        self.__nombre = valor.strip()

    @property
    def promedio(self):
        return self.__promedio

    @promedio.setter
    def promedio(self, valor):
        if not isinstance(valor, (int, float)) or not (0 <= valor <= 5):
            raise ValueError(f"El promedio debe estar entre 0 y 5 (recibido: {valor}).")
        self.__promedio = float(valor)

    def __repr__(self):
        return f"Estudiante(nombre='{self.__nombre}', promedio={self.__promedio})"
```

```python
e4 = EstudianteV4("Ana", 4.2)
print(e4)
```

```python
# Interfaz limpia: se usa como un atributo normal
e4.promedio = 3.8
print("Promedio actualizado:", e4.promedio)
```

```python
# Las mismas restricciones que V3, pero con sintaxis de atributo
try:
    e4.promedio = 9.9
except ValueError as ex:
    print("[promedio 9.9]", ex)
```

```python
try:
    e4.nombre = None
except ValueError as ex:
    print("[nombre None]", ex)
```

```python
# El atributo privado sigue siendo inaccesible directamente
try:
    print(e4.__promedio)
except AttributeError as ex:
    print("[acceso directo __promedio]", ex)
```

## Resumen comparativo — Encapsulamiento

| Versión | Sintaxis del atributo | Acceso externo | Validación posible |
|---|---|---|---|
| **V1** | `self.x` | Libre, sin restricción | No |
| **V2** | `self._x` | Libre (por convención no se toca) | No |
| **V3** | `self.__x` + `get_x()` / `set_x()` | Bloqueado por name mangling | Sí, dentro del setter |
| **V4** | `self.__x` + `@property` | Parece público, está protegido | Sí, automática en el setter |

> **Regla práctica:** usa `@property` (V4) cuando necesites controlar el acceso a un atributo. Es la forma pythónica: la interfaz es limpia y la validación es automática.

## Introducción a polimorfismo

```python
def sumar(L):
    # Retorna la suma de elementos de una lista no vacia
    suma = L[0]
    for i in range(1,len(L)):
        suma += L[i] 
    return suma 
```

```python
sumar([1,2,3]), sumar(['a', 'b', 'c']), sumar([Fraccion(2,3), Fraccion(2,3), Fraccion(2,3)])
```

## ¿Qué está pasando en `sumar`?

La función `sumar` no sabe el tipo de los elementos: solo usa `+=`. Funciona con enteros, strings y `Fraccion` porque los tres tienen definido el operador `+` (`__add__`). Esto se llama **duck typing**: *"si camina como pato y grazna como pato, es un pato"*. Python no verifica el tipo; verifica que el objeto tenga el comportamiento requerido.

## Polimorfismo con herencia

La forma más clásica: una clase base define la **interfaz** (qué métodos deben existir) y cada subclase la implementa a su manera.

```python
import math

class Figura:
    def area(self):
        raise NotImplementedError("Cada figura debe implementar area()")

    def perimetro(self):
        raise NotImplementedError("Cada figura debe implementar perimetro()")

    def __repr__(self):
        return f"{self.__class__.__name__} | área={self.area():.2f} | perímetro={self.perimetro():.2f}"


class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * self.radio ** 2

    def perimetro(self):
        return 2 * math.pi * self.radio


class Rectangulo(Figura):
    def __init__(self, base, altura):
        self.base   = base
        self.altura = altura

    def area(self):
        return self.base * self.altura

    def perimetro(self):
        return 2 * (self.base + self.altura)


class Triangulo(Figura):
    def __init__(self, a, b, c):
        self.a, self.b, self.c = a, b, c

    def area(self):
        s = self.perimetro() / 2
        return math.sqrt(s * (s-self.a) * (s-self.b) * (s-self.c))

    def perimetro(self):
        return self.a + self.b + self.c
```

```python
figuras = [Circulo(5), Rectangulo(4, 6), Triangulo(3, 4, 5)]

for f in figuras:
    print(f)
```

El mismo bucle llama a `area()` y `perimetro()` sobre objetos de tipos distintos. Cada clase ejecuta **su propia versión** del método: eso es polimorfismo.

La función `__repr__` está definida **una sola vez** en `Figura` y funciona para todas las subclases porque internamente llama a `self.area()` y `self.perimetro()`, que se resuelven en tiempo de ejecución según el tipo real del objeto.

## Polimorfismo en funciones

Una función que recibe cualquier `Figura` sin importar cuál:

```python
def figura_mayor(figuras):
    """Retorna la figura con mayor área de una lista."""
    return max(figuras, key=lambda f: f.area())

def area_total(figuras):
    """Suma el área de todas las figuras de la lista."""
    return sum(f.area() for f in figuras)

print("Mayor área:", figura_mayor(figuras))
print("Área total:", round(area_total(figuras), 2))
```

## `isinstance` — verificar el tipo en tiempo de ejecución

A veces una función necesita comportarse diferente según el tipo concreto del objeto. `isinstance` permite comprobarlo sin romper el polimorfismo:

```python
def describir(figura):
    if isinstance(figura, Circulo):
        print(f"Círculo de radio {figura.radio}")
    elif isinstance(figura, Rectangulo):
        print(f"Rectángulo de {figura.base}×{figura.altura}")
    elif isinstance(figura, Triangulo):
        print(f"Triángulo de lados {figura.a}, {figura.b}, {figura.c}")

    # Esto funciona para cualquier Figura sin importar cuál:
    print(f"  → Es una Figura: {isinstance(figura, Figura)}")
    print(f"  → área={figura.area():.2f}\n")

for f in figuras:
    describir(f)
```

## Polimorfismo en métodos especiales

Python usa métodos especiales (`__add__`, `__len__`, `__str__`, …) para que los operadores y funciones built-in funcionen con cualquier clase. Ya lo vimos con `Fraccion`; aquí otro ejemplo: una clase `Vector` que soporta `+`, `len` y comparación con `==`.

```python
class Vector:
    def __init__(self, *componentes):
        self._v = list(componentes)

    def __len__(self):
        return len(self._v)

    def __add__(self, other):
        if len(self) != len(other):
            raise ValueError("Los vectores deben tener la misma dimensión.")
        return Vector(*(a + b for a, b in zip(self._v, other._v)))

    def __eq__(self, other):
        return self._v == other._v

    def __repr__(self):
        return f"Vector{tuple(self._v)}"


v1 = Vector(1, 2, 3)
v2 = Vector(4, 5, 6)

print(v1 + v2)          # Vector(5, 7, 9)   — usa __add__
print(len(v1))          # 3                 — usa __len__
print(v1 == v2)         # False             — usa __eq__
print(v1 == Vector(1, 2, 3))  # True

# sumar también funciona con vectores gracias a duck typing
print(sumar([v1, v2, Vector(0, 0, 0)]))
```

---
## Resumen — Polimorfismo

| Mecanismo | Cómo funciona | Ejemplo en esta clase |
|---|---|---|
| **Duck typing** | Si el objeto tiene el método, funciona | `sumar` con `int`, `str`, `Fraccion`, `Vector` |
| **Herencia + override** | La subclase redefine el método de la clase base | `Circulo.area()`, `Rectangulo.area()`, `Triangulo.area()` |
| **Métodos especiales** | `__add__`, `__len__`, `__eq__`… integran la clase con operadores Python | `Fraccion`, `Vector` |
| **`isinstance`** | Permite actuar diferente según el tipo concreto sin romper la abstracción | `describir(figura)` |

> **Clave:** el polimorfismo permite escribir código **genérico** que funciona con tipos distintos, siempre que compartan la interfaz esperada.
