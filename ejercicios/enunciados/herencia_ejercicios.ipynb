## 🚗 Ejercicio 1: Vehículos

### Clase base: `Vehiculo`

### Clase `Moto(Vehiculo)`:

- `tipo`: tipo de moto (ej: `'deportiva'`, `'crucero'`)

## Métodos:
- `hacer_caballito()`: retorna `'¡Caballito!'` si `velocidad > 30`, de lo contrario `'Necesitas más velocidad'`

---

## Prueba:

---

### Clase base: `Figura`

## Atributos:
- `color`

## Métodos:
- `area()`: lanzar `NotImplementedError('Implementar en subclase')`
- `perimetro()`: lanzar `NotImplementedError('Implementar en subclase')`
- `__str__`: retorna `"Figura de color {color}"`
- `descripcion()`: retorna `"Soy una {tipo} de color {color} con área {area:.2f} y perímetro {perimetro:.2f}"`, donde `tipo` es el nombre de la clase

---

### Clase `Circulo(Figura)`:

## Atributos adicionales:
- `radio`

## Métodos:
- `area()`: retorna $\pi \times r^2$
- `perimetro()`: retorna $2 \times \pi \times r$
- Sobreescribir `__str__`: retorna `"Círculo de radio {radio} y color {color}"`

---

### Clase `Triangulo(Figura)`:

## Atributos adicionales:
- `base`, `altura`
- `lado1`, `lado2`, `lado3`

## Métodos:
- `area()`: retorna `(base × altura) / 2`
- `perimetro()`: retorna la suma de los tres lados
- Sobreescribir `__str__`: retorna `"Triángulo de base {base} y color {color}"`

---

## Prueba:

```python
import math

c = Circulo('rojo', 5)
print(c)                    # Círculo de radio 5 y color rojo
print(round(c.area(), 2))   # 78.54
print(c.descripcion())      # Soy una Circulo de color rojo con área 78.54 y perímetro 31.42

t = Triangulo('azul', 4, 3, 5, 4, 3)
print(t.area())             # 6.0
print(t.perimetro())        # 12
print(t.descripcion())      # Soy una Triangulo de color azul con área 6.00 y perímetro 12.00
```

### Clase `Pila`:

- `__len__`: retorna la cantidad de elementos
- `__str__`: retorna `"Pila (tope → base): {elementos_de_tope_a_base}"`
