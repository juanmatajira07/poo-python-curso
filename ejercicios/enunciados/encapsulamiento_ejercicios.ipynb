# Ejemplo de lo que hace `@property` y `@setter`

En el ejemplo siguiente se presenta un clase que es cada vez mГЎs compleja y los cambios que se presentan en dicha clase para entender atributos privados `__atributo` y como se pueden manejar con mГ©todos de la clase. 

## Un ejemplo usual de un atributo pГєblico y uno privado

````python
class Usuario:
    def __init__(self, nombre, documento):
        self.nombre = nombre
        self.__documento = documento

    def __repr__(self):
        return f'Nombre: {self.nombre} | Documento: {self.__documento}'
````

````python
U1 = Usuario('Arley', 123456)
U1
````

````python
U1.nombre = 'Carlos'
U1
````

````python
U1.__documento
````

## Poniendo mГ©todos para leer la informaciГіn y cambiarla

````python
class Usuario:
    def __init__(self, nombre, documento):
        self.nombre = nombre
        self.__documento = documento

    def get_documento(self):
        return self.__documento
    
    def set_documento(self, valor):
        self.__documento = valor

    def __repr__(self):
        return f'Nombre: {self.nombre} | Documento: {self.__documento}'
````

````python
U2 = Usuario('Arley', 123456)
U2
````

````python
print(U2.get_documento())
U2.set_documento(9876)
U2
````

## ВїQuГ© hace @property y @setter?

1. @property me permite tomar un mГ©todo y ejecutar `instancia.metodo` para que sea de lectura
2. @setter me permite tomar un mГ©todo y ejecutarlo como `instancia.metodo = valor`

````python
class Usuario:
    def __init__(self, nombre, documento):
        self.nombre = nombre
        self.__documento = documento

    @property                   # permite instancia.documento1 y retorna self._documento
    def documento1(self):
        return self.__documento
    
    @documento1.setter          # permite instancia.documento1 = valor y lo guarda en self.__documento
    def documento1(self, valor):
        self.__documento = valor

    def __repr__(self):
        return f'Nombre: {self.nombre} | Documento: {self.__documento}'
````

````python
U3 = Usuario('Arley', 123456)
U3
````

````python
print(U3.documento1) # Se llama como un atributo pГєblico usual
U3.documento1 = 9876 # Se asigna un nuevo valor como un atributo pГєblico usual
U3
````

## Finalmente

- ВїQuГ© sucede si `documento1` se llama `documento`?

````python
class Usuario:
    def __init__(self, nombre, documento):
        self.nombre = nombre
        self.__documento = documento

    @property
    def documento(self):
        return self.__documento
    
    @documento.setter
    def documento(self, valor):
        self.__documento = valor

    def __repr__(self):
        return f'Nombre: {self.nombre} | Documento: {self.__documento}'
````

````python
U4 = Usuario('Arley', 123456)
U4
````

````python
# La sintГЎxis se vuelve mГЎs sencilla de manejar en la clase
print(U4.documento) # Se llama como un atributo pГєblico usual
U4.documento = 9876 # Se asigna un nuevo valor como un atributo pГєblico usual
U4
````
